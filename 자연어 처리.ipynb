```python
import nltk
```

```python
# 분석 대상 문장
sentence = """At eight o'clock' on Thursday morning
Arthur didn't feel very good."""
```

```python
# 자연어 처리용 사전이 기본적으로 설치되지 않음.
# 사전이 설치되지 않으면 오류 발생
tokens = nltk.word_tokenize(sentence)
```

```python
# 대상 영어 문장 요소별 추출
tokens = nltk.word_tokenize(sentence)
tokens
```

영어 및 그와 유사한 언어들은 요소(token)의 추출이 띄어쓰기만을 기준으로 해도 상당 수준 가능

```python
tokens_from_space_split = sentence.split(' ')
print(tokens_from_space_split)
```

NLTK 사용 시와 띄어쓰기만을 사용한 경우 요소 추출 결과 비교 

```python
common = list(set(tokens)&set(tokens_from_space_split))
print(common)
```

```python
diff = list(set(tokens)-set(tokens_from_space_split))
print(diff)
```

띄어쓰기만으로도 상당 수준 되긴 하지만 NLTK의 경우는 다음과 같이 문장의 요소를 추출했다.

1. didn't --> [did, n't]
2. morning\nArthur --> [morning, Arthur]
3. good. --> [good, .]

태그 생성

```python
tagged = nltk.pos_tag(tokens)
print(tagged)
```
