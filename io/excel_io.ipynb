## 엑셀 파일 읽기

```python
프레임목록 = {}
for 시즌 in range(2013, 2016):
    프레임 = pd.read_excel(join(데이터경로, 'NC Dinos {}.xlsx').format(시즌))
    프레임목록[시즌] = 프레임
```

하나의 엑셀 파일에 여러 시트로 저장하기

```python
NC시즌 = pd.ExcelWriter('NC Dinos.xlsx')

for 시즌, 프레임 in 프레임목록.items():
    프레임.to_excel(NC시즌, str(시즌))

NC시즌.save()
```

# 한국수출입은행 (공공데이터) 엑셀 파일 처리

```python
경로 = '../data/koreaexim/international/Global+Market+Daily+Table%28%EA%B0%92%EB%B3%B5%EC%82%AC%29_0803.xls'
```

파일 경로에 포함된 이상한 기호들의 정체

```python
from urllib.parse import unquote_plus as url디코드
url디코드(경로)
```

액셀 파일을 읽어보니 위쪽 빈줄이 많다.

```python
pd.read_excel(경로)
```

위의 3줄은 건너뛰기로 한다.

```python
pd.read_excel(경로, skiprows=3)
```

열은 액셀의 C열부터 K열까지 가져오도록 한다.

```python
pd.read_excel(경로, skiprows=3, parse_cols="C:K")
```

표가 여러 개라 열제목은 일단 없음으로 설정해야겠다.

```python
국제0803 = pd.read_excel(경로, skiprows=3, parse_cols="C:K", header=None)
국제0803
```

빈 줄은 삭제하자

```python
국제0803 = 국제0803.dropna(how='all')
국제0803
```

빈 줄이 삭제되면서 색인 번호가 고르지 않아졌다.

```python
try: 
    국제0803.ix[8] # KeyError
except KeyError:
    print('그런 색인 없어요!')
```

색인을 재설정하자

```python
국제0803 = 국제0803.reset_index(drop=True)
국제0803
```

각 표를 각각의 데이터프레임으로 잘라내자

```python
국제0803.iloc[0:8]
```

범위 목록을 만들어서 한번에 처리하자

```python
표정보목록 = [('equity', 0,8), 
          ('fixed income',8,18), 
          ('fx',16,22), 
          ('commodity',24,29), 
          ('cds',29,44)]
```

```python
국제0803_표사전 = {}
for 표정보 in 표정보목록:
    제목, 시작, 종료 = 표정보
    국제0803_표사전[제목] = 국제0803.iloc[시작:종료]
```

```python
국제0803_표사전['cds']
```

색인범위를 자동으로 구할 수도 있을까?

```python
색인시작목록 = list(국제0803[국제0803.notnull().sum(1) == 1].index)
색인시작목록
```

```python
색인종료목록 = 색인시작목록[1:]
색인종료목록.append(len(국제0803))
색인종료목록
```

제목 추출

```python
제목목록 = [국제0803.iloc[i,0].lower() for i in 색인시작목록]
제목목록
```

```python
표정보목록 = list(zip(제목목록, 색인시작목록, 색인종료목록))
표정보목록
```
